# Lond-Short Term Memory 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np
```

```python
X = np.array([
    [[1], [2], [3]],
    [[2], [3], [4]],
    [[3], [4], [5]]
])
```

```python
y = np.array([4, 5, 6])
```

```python
model = Sequential()
```

```python
X=X.reshape((3,3,1))
```

```python
model.add(LSTM(50, input_shape=(3,1)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=100, verbose="auto")
```

```python
test_input=np.array([[[5],[6],[7]]])
prediction=model.predict(test_input,verbose=0)
```

```python
print("Predicted Next Number :",prediction[0][0])
```
