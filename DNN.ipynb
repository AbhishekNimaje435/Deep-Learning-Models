# Deep Neural Network (Diabetes Diagnosis Test)

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Input: [Glucose, BMI]
X = np.array([
    [90, 22],
    [85, 20],
    [88, 23],
    [150, 35],
    [160, 38],
    [170, 40]
])
```

```python
# Output: Diabetes (0 = No, 1 = Yes)
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
model = Sequential()
```

```python
model.add(Dense(4, activation='relu', input_shape=(2,)))
model.add(Dense(1, activation='sigmoid'))
```

```python
 model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(X, y, epochs=200,verbose=0)
```

```python
# New patient
patient = np.array([[120, 28]])

prediction=model.predict(patient)
```

```python
prediction
```

```python
if prediction >= 0.5:
    print("Diabetes")
else:
    print("No Diabetes")
```
