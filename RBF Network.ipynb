# Radial Based Function Network (RBF Network)

```python
import numpy as np
```

```python
# Training data
X = np.array([1,2,6,7])
```

```python
# RBF centers
centers = np.array([1.5, 6.5])
sigma = 1
```

```python
def rbf(x, c, sigma):
    return np.exp(-(x - c)*2 / (2 * sigma*2))
```

```python
# Hidden layer output
H = np.array([[rbf(x, c, sigma) for c in centers] for x in X])
print("Hidden layer:\n", H)
```

```python
# Output layer weights
weights = np.array([1, 2])
```

```python
# Prediction
y = H @ weights
prediction = (y > 0.5).astype(int)
```

```python
print("RBF output:", y)
print("Predicted class:", prediction)
```
